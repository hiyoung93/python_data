 ## concat 
 - index를 기준으로 데이터를 합친다.
 - pd.concat([변수명,변수명,변수명], **keys** = ['변수를', '그룹명으로', '짝지어줌'])
 - key = level = groupby 이름으로 만들어줌
 - pd.concat([변수명,변수명,변수명], **axis = 1**); <br>axis 는 index를 기준으로 data를 병합,값이 없는 부분은 NaN으로 저장
 - pd.concat([변수명,변수명,변수명], axis = 1, **join='inner'**); <br>공통되지 않는 값을 가지고 있는 부분은 버려짐

# concat& merge연습하기

```python
df1 = pd.DataFrame({'A': ['A0','A1','A2','A3'],
                    'B': ['B0','B1','B2','B3'],
                    'C': ['C0','C1','C2','C3'],
                    'D': ['D0','D1','D2','D3']},
                  index =[0, 1, 2, 3])
df2 = pd.DataFrame({'A': ['A4','A5','A6','A7'],
                    'B': ['B4','B5','B6','B7'],
                    'C': ['C4','C5','C6','C7'],
                    'D': ['D4','D5','D6','D7']},
                  index =[4, 5, 6, 7])

df3 = pd.DataFrame({'A': ['A8','A9','A10','A11'],
                    'B': ['B8','B9','B10','B11'],
                    'C': ['C8','C9','C10','C11'],
                    'D': ['D8','D9','D10','D11']},
                  index =[8, 9, 10, 11])
```

```python
result = pd.concat([df1, df2, df3])
# 열방향으로 합치기 
# pd.concat([변수명,변수명0,변수명])
result
```

```python
result = pd.concat([df1,df2,df3], keys = ['x','y','z'])
# concat은 key 옵션으로 구분 가능, 
# key 지정된 구분은 다중 index가 되서 level생성됨
result
```

```python
result.index.get_level_values(0)
```

```python
df4 = pd.DataFrame({'B': ['B2','B3','B6','B7'],
                    'D': ['D2','D3','D6','D7'],
                    'F': ['F2','F3','F6','F7']},
                    index = [2,3,6,7])
result = pd.concat([df1,df4], axis = 1)
# axis=1 은 data를 병합해준다. 값이 없는 부분은 NaN으로 저장됨
# pd.concat([변수명,변수명], axis=1)
```

```python
result
```

```python
result = pd.concat([df1,df4], axis=1, join='inner')
# join='inner'로 공통되지 않는 부분을 가지고 있는 부분은 버려진다
# pd.concat([변수명,변수명], axis=1, join='inner')
result
```

## merge - columns를 기준으로 변경됨
- pd.merge(변수명, 변수명, on='공통columns명') -- <br> 공통 columns name을 기준으로 on옵션을 하면 **같은 columns name의 데이터 만** 가지고 온다

- pd.merge(변수명, 변수명, how= 'left', on='공통columns명') -- <br> 공통 columns name을 기준으로 on 옵션을 한 후, how옵션으로 **왼쪽이 기준**으로 간다.

- pd.merge(변수명, 변수명, how= 'right', on='공통columns명') -- <br> 공통 columns name을 기준으로 on 옵션을 한 후, how옵션으로 **오른쪽이 기준**으로 간다.

- pd.merge(변수명, 변수명, how= 'outer', on='공통columns명') -- <br> 공통 columns name을 기준으로 on 옵션을 한 후, how옵션으로 **합집합**으로 간다. <br> 공통된 요소가 아닌 곳은 **NaN**가 된다.

- pd.merge(변수명, 변수명, how= 'inner', on='공통columns명') -- <br> 공통 columns name을 기준으로 on 옵션을 한 후, how옵션으로 **교집합**으로 간다. <br> **공통된 요소 만** 가지고 온다


```python
pd.merge(left, right, on='key')
# pd.merge(변수명, 변수명, on='공통columns명')
# 같은 columns name의 데이터만 가지고 온다
```

```python
pd.merge(left, right, how='left',on='key')
# pd.merge(left, right, how='left',on='공통columns명')
# left를 기준으로 합쳐진다
# 없는 값은 NaN
```

```python
pd.merge(left, right, how='right', on='key')
# pd.merge(lefg, right, how='right', on='공통columns명')
# right를 기준으로 합쳐진다  
# 없는 값은 Na
```

```python
pd.merge(left, right, how='inner', on='key')
# pd.merge(변수명, 변수명, how='inner', on='공통columns명')
# columns name이 같은 것만 합쳐진다
```

```python
pd.merge(left, right, how='outer',on='key')
# pd.merge(변수명, 변수명, how='outer',on='공통columns명')
# columns name을 전부 합친다
# 없는 값은 NaN
```
