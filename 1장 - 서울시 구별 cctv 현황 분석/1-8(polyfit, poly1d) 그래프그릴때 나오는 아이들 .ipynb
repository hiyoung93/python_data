```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
t = np.arange(0, 10, 0.01)
y = 3*t + 5
# 기울기가 3, y절편 5
```

```python
plt.figure(figsize=(12,8))
plt.plot(t, y)
plt.show()
```

```python
np.random.seed(3)
y_noise = y +np.random.randn(len(y))
# 변수 = y + 랜덤(y의 길이만큼)
plt.figure(figsize=(12,8))
plt.plot(t, y_noise)
plt.show()
```

```python
fp1 =np.polyfit(t, y_noise, 1)
fp1
```

 np.polyfit(변수, 변수, 함수의 차수)

- 직선을 구성하는 것은 기울기와 절편
- polyfit명령으로 쉽게 찾을 수 있다.
- polyfit의 세번째 값은 함수의 차수이며, 2를 넣으면 2차식 계수를 찾아달라고 한다

```python
f1 = np.poly1d(fp1)
f1
```

```python
np.poly1d([1,1]) + np.poly1d([1, -1])
```

```python
np.poly1d([1,1])*np.poly1d([1,-1])
```

```python
np.poly1d([1,-2,1])/np.poly1d([1,-1])
```

### np.poly1d(np.polyfit(변수, 변수, 함수의 차수))
- polynomial class 로 만들어주는 아이
- 다항식연산과 같음
- 다항식 연산이 가능한 numpy클래스중 하나이다.


```python
plt.figure(figsize=(12,8))
plt.plot(t, y_noise, label='noise', color='y')
plt.plot(t, y, ls='dashed', lw=3, color='b', label='original')
plt.plot(t, f1(t), lw=2, color='r', label='polyfit')
plt.grid()
plt.legend()
plt.show()
```
