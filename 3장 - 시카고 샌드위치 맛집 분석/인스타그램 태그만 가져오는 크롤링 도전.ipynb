```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import time
import requests
from selenium import webdriver
```

인스타그램은 Beautifulsoup 가지고 절대 크롤불가

```python
# 함수 생성해서 requests.get으로 url얻기
def crawl(url):
    data = requests.get(url)
    print(data, url)
    return data.content
```

```python
# url 나누기
url_base = 'https://www.instagram.com'
url_sub = '/explore/tags/doll/'
url = url_base + url_sub
```

```python
# 잘 실행 되는지 확인하기
pageString = crawl(url)
```

```python
# 크롬 드라이버로 크롬 열기
driver = webdriver.Chrome('chromedriver.exe')
```

```python
# 열린 크롬 창에서 url 실행하기
driver.get(url)
```
